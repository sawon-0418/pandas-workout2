```python
import pandas as pd
from pandas import Series, DataFrame

df = pd.read_csv('../data/linux-etc-passwd.txt', 
                 sep=':', comment='#', header=None,
                names=['username', 'password', 'userid', 'groupid', 'name', 'homedir', 'shell'])
df
#리눅스에서 etc-passwd 파일은 사용자 계정 정보를 담고 있는 파일이다.
#이 파일을 csv파일로 읽어들여 DataFrame으로 만든다.
#이때, 구분자로 : 를 사용하고, #으로 시작하는 것은 주석으로 처리하여 무시한다.
#또한, 헤더가 없으므로 header=None으로 지정하고, 컬럼 이름을 names로 지정한다.
# 아래는 파일 내용 예시이다.
# # This is a comment
# # You should ignore me
# root:x:0:0:root:/root:/bin/bash

#원래 csv파일은 기본적으로 ,로 구분하지만, sep 파라미터를 사용하여 다른 구분자를 지정할 수 있다.
```
