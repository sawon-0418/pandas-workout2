# Beyond 2

Unix systems typically reserve user IDs below 1000 to special accounts.  Show the non-special usernames in this `passwd` file.

번역: 유닉스 시스템에서는 일반적으로 사용자 ID 1000 미만을 특수 계정에 할당합니다. 이 passwd 파일에서 특수 계정이 아닌 사용자 이름들을 보여주세요.

```python
df = pd.read_csv('../data/linux-etc-passwd.txt',
                 sep=':', comment='#', header=None,
                names=['username', 'password', 'userid', 'groupid', 'name', 'homedir', 'shell'])

df[df['userid'] > 1000]
# >= 이 아닌, >가 맞음.


# 대부분의 현대 리눅스 배포판(예: Debian/Ubuntu, 최신 Fedora 등)에서는 UID 1000부터 일반(사람) 사용자 계정을 할당합니다.

# 0: root
# 1–99: 전통적 시스템 계정
# 100–999: 서비스/시스템 계정
# 1000 이상: 일반 사용자
# 예외: 오래된 Red Hat 계열은 500부터 시작했고, 일부 시스템은 정책을 바꿀 수 있습니다. 따라서 “보통 1000은 일반 사용자”가 맞지만, 절대적인 규칙은 아니며 /etc/login.defs 의 UID_MIN 값을 확인하면 확실합니다.
```

# Beyond 3

Immediately after logging into a Unix system, a command interpreter, known as a "shell," fires up. What are the different shells in this file?


번역: 유닉스 시스템에 로그인하자마자 “셸(shell)”이라 불리는 명령 해석기가 실행됩니다. 이 파일에 나타나는 서로 다른 셸은 무엇인가요?

```python
# The `unique` method does this for us, returning a NumPy array -- but Pandas offers its own
# drop_duplicates method, which keeps us with a series
# 번역: unique 메서드는 이를 수행하여 NumPy 배열을 반환합니다. 
# 그러나 Pandas는 자체 drop_duplicates 메서드를 제공하며, 이를 사용하면 Series 형태를 유지할 수 있습니다.

df['shell'].drop_duplicates()

#drop_duplicates() : 중복된 값을 제거하고 고유한 값만 남깁니다.
#넘파이 에서는 unique() 메서드를 사용하며, 넘파이 어레이를 반환한다.
```
