```python
import pandas as pd
from pandas import Series, DataFrame

filename = '../data/cities.json'
df = pd.read_json(filename)

df
```

# Beyond 1

Convert the `growth_from_2000_to_2013` column into a floating-point number. Then find the mean and median changes in city size between 2000 and 2013.

growth_from_2000_to_2013 열을 부동소수(float)형으로 변환하세요. 그런 다음 2000년부터 2013년 사이 도시 규모 변화의 평균과 중앙값을 구하세요.

```python
# Remove trailing %
df['growth_from_2000_to_2013'] = df['growth_from_2000_to_2013'].str.rstrip('%')
#원래 n% 였던 값을, rstrip('%')를 통해 %를 제거
# .str 은 pandas 문자열 접근자
#rstrip('%')는 각 문자열의 오른쪽 끝에서 %문자만 제거


# Find empty strings, and turn to 0
df.loc[df['growth_from_2000_to_2013'] == '', 'growth_from_2000_to_2013'] = '0'
#    "city": "Jurupa Valley", 
#         "growth_from_2000_to_2013": "", 
#         "latitude": 33.9971974, 
#         "longitude": -117.4854802, 
#         "population": "98030", 
#         "rank": "300", 
#         "state": "California" 이처럼 gorwth_from_2000_to_2013 값이 빈 문자열인 경우가 있음 그것을 0으로 바꿈

# Turn to float, return mean + median
df['growth_from_2000_to_2013'].astype(float).describe()[['mean', '50%']]
# astype(float) : 문자열을 실수로 변환
# describe() : 요약 통계량 산출
# [['mean', '50%']] : mean과 50%(median) 값만 추출

```

# Beyond 2

How many cities had positive growth in this period, and how many had negative growth?


```python
df['growth_from_2000_to_2013'] = df['growth_from_2000_to_2013'].astype(float)
#위 코드를 작성하지 않으면 문자열로 존재하기에 bins 가 제대로 작동할 수 없음

pd.cut(df['growth_from_2000_to_2013'],
        bins=[df['growth_from_2000_to_2013'].min(), 0, df['growth_from_2000_to_2013'].max()],
        include_lowest=True,
        labels=['-', '+']).value_counts()
```

# Beyond 3

Find the cities that whose latitude is more than two standard deviations away from the mean.

번역: 위도가 평균에서 표준편차의 2배를 초과하여 벗어난 도시들을 찾아라.

```python
df[(df['latitude'] > df['latitude'].mean() + 2*df['latitude'].std()) |
   (df['latitude'] < df['latitude'].mean() - 2*df['latitude'].std()) ]

# 위 코드는 위도(latitude) 값이 평균에서 표준편차의 2배 이상 벗어나는 도시들을 찾는 코드
df.loc[(df['latitude'] > df['latitude'].mean() + 2*df['latitude'].std()) |
         (df['latitude'] < df['latitude'].mean() - 2*df['latitude'].std()), 
         'city']
#위 코드처럼 하면 도시 이름만 추출 가능
```
