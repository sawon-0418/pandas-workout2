```python
import pandas as pd
from pandas import Series, DataFrame

s = pd.read_csv('../data/nyc-temps.txt').squeeze()
```

# Beyond 1

Set all of the odd temperatures to the mean of all temperatures
모든 홀수 온도 값을 전체 온도의 평균으로 설정하세요

```python
df = DataFrame({'temp': s*1.0, 
                'hour': [0,3,6,9,12,15,18,21] * 91})
# temp에 s*1.0을 곱하는 이유는 int형을 float형으로 바꾸기 위해서
df.loc[
    df['temp']%2 == 1, 
    'temp'
] = df['temp'].mean()
# temp 속성의 값이 2로 나누었을 때 나머지가 1인(즉, 홀수인) 행을 찾아서 
# 그 행의 temp 값을 temp 속성의 평균값으로 바꿈
df.describe()
```

# Beyond 2

Set the even temperatures at hours 9 and 18 to 3
9시와 18시의 짝수 온도 값을 3으로 설정해라.

```python
df = DataFrame({'temp': s, 
                'hour': [0,3,6,9,12,15,18,21] * 91})
df.loc[
    (df['hour'].isin([9, 18])) & (df['temp'] % 2 == 0),
    'temp'] = 3

df

# hour 속성의 값에 9 또는 18이 존재하는 행의 온도를 3으로 바꿈
df.describe()
```

# Beyond 3

If the hour is odd, then set the temperature to 5

시간이 홀수라면, 온도를 5로 설정해라

```python
df = DataFrame({'temp': s, 
                'hour': [0,3,6,9,12,15,18,21] * 91})
df.loc[
    df['hour']%2 == 1, 
    'temp'
] = 5
#hour 컬럼의 값이 2로 나누었을 때 나머지가 1인(즉, 홀수인) 행을 찾아서
# 그 행의 temp 값을 5로 바꿈


df.describe()
```
